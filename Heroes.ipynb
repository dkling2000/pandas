```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
# File to Load (Remember to Change These)
purchase_data = os.path.join("resources","purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(purchase_data)
```

## Player Count

* Display the total number of players


```python
count = len(purchase_data_df["SN"].value_counts())
print("Total number of players: " + str(count))

```

## Purchasing Analysis (Total)

```python
uitems = len(purchase_data_df["Item ID"].unique())
avgprice = purchase_data_df["Price"].mean()
#round with dollar sign to 2 decimL places
totpur = len(purchase_data_df["Purchase ID"])
#add $ to format
trev = purchase_data_df['Price'].sum()

summary_df = pd.DataFrame({"Number of Unique Item": uitems,"Average Price": avgprice, 
                           "Number of Purchases": totpur, "Total Revenue": trev}, index = [0])

#summary_df["avgprice"] = summary_df["avgprice"].map('${:,.2f}'.format)
#summary_df["trev"] = summary_df["trev"].map("${:,.2f}".format)
summary_df
#df['cost'] = df['cost'].map('${:,.2f}'.format)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
males = (purchase_data_df["Gender"] == "Male").sum()
females = (purchase_data_df["Gender"] == "Female").sum()
other = (purchase_data_df["Gender"] == "Other / Non-Disclosed").sum()
total = males + females + other
pctmales = round((males/total * 100),2)
pctfemales = round((females/total * 100),2)
pctother = round((other/total * 100),2)

gensum = pd.DataFrame({"Total": [males, females, other],"Percentage of Players": 
                       [pctmales,pctfemales,pctother]}, index=["Males","Females","Other / Non-Disclosed"])
gensum
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender



* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
fem_avg_pri = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", "Price"].mean()
male_avg_pri = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", "Price"].mean()
oth_avg_pri = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", "Price"].mean()
fem_tot_pri = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", "Price"].sum()
male_tot_pri = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", "Price"].sum()
oth_tot_pri = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", "Price"].sum()
fem_uname = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", "SN"].nunique()
male_uname = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", "SN"].nunique()
oth_uname = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", "SN"].nunique()
fem_avg_pper = fem_tot_pri/fem_uname
male_avg_pper = male_tot_pri/male_uname
oth_avg_pper = oth_tot_pri/oth_uname



summary_df = pd.DataFrame({"Gender": ["Females","Males","Other / Non-Disclosed"],
                           "Purchase Count":[females,males,other], 
                           "Average Purchase Price":[fem_avg_pri,male_avg_pri,oth_avg_pri],
                           "Total Purchase Value":[fem_tot_pri,male_tot_pri,oth_tot_pri],
                            "Avg Total Per Person":[fem_avg_pper,male_avg_pper,oth_avg_pper]}, index=['Female','Male','Other / Non-Disclosed'])

summary_df



```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,1000]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-40','40+']

purchase_data_df['Total Count']=pd.cut(purchase_data_df['Age'],bins,labels=labels)

new_purch =purchase_data_df.drop_duplicates(['SN'], keep = 'last')

pplayer = round((1/new_purch['SN'].count() *100),2)
new_purch['Percentage of Players']= pplayer

age_group = new_purch.groupby('Total Count', as_index = 'Total Count').agg({'Total Count': "count",'Percentage of Players': "sum"})

age_group

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9,14,19,24,29,34,39,1000]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-40','40+']

purchase_data_df['Purchase Count']=pd.cut(purchase_data_df['Age'],bins,labels=labels)
age_analysis_df = purchase_data_df

avg_pur_total = age_analysis_df.groupby(['Purchase Count']).sum()["Price"].rename('Total Purchase Value')

norm_total = avg_pur_total/age_group['Total Count']
age_analysis_df['Average Purchase Price'] =age_analysis_df['Price']
age_analysis_df['Total Purchase Value'] =age_analysis_df['Price']

#age_analysis_df['Avg Total per Person'] = age_analysis_df['Total Purchase Value'].sum()/purchase_data_df['Total Count'].count()


group_age_analysis = age_analysis_df.groupby('Purchase Count').agg({'Purchase Count': "count",
                                            'Average Purchase Price': 'mean','Total Purchase Value':'sum'})
group_age_analysis ['Avg Total Purchase per Person'] = norm_total = avg_pur_total/age_group['Total Count']
group_age_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spend_df = purchase_data_df

top_spend_df['Average Purchase Price'] = top_spend_df["Price"]
top_spend_df['Total Purchase Price'] = top_spend_df["Price"]
top_spend_df['Purchase Count'] = top_spend_df ['SN']
name_group = top_spend_df.groupby('SN').agg({'Purchase Count': "count",'Average Purchase Price':'mean','Total Purchase Price':'sum'})
name_group = name_group.sort_values('Total Purchase Price', ascending = False)
name_group.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = purchase_data_df

items_df['Purchase Count'] = items_df['SN']
items_df['Item Price'] = items_df['Price']
items_df['Total Purchase Value'] = items_df['Price']

items_sum_df =items_df.groupby(['Item ID','Item Name']).agg({'Purchase Count':'count','Item Price':'mean','Total Purchase Value':'sum'})
items_sum_df = items_sum_df.sort_values('Purchase Count', ascending = False)

items_sum_df.head ()
```
